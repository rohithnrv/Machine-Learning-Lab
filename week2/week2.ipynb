```python
#12. Write a function to find min and max values of any column

# Step 1: Define the function
def find_min_max(df, weight):
    if weight in df.columns:
        min_val = df[weight].min()
        max_val = df[weight].max()
        return min_val, max_val
    else:
        return f"Column '{weight}' not found in the DataFrame."

# Step 2: Call the function with your DataFrame and column name
min_price, max_price = find_min_max(df, 'Price in thousands')

# Step 3: Print results
print("Minimum Price:", min_price)
print("Maximum Price:", max_price)


def min_max(df,column):
  min_val=df[column].min()
  max_val=df[column].max()
  return min_val,max_val

column='Sales in thousands'
print(min_max(df,column))
```
